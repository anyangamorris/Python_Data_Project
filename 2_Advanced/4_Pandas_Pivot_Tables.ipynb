```python
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
dataset = load_dataset ('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

```

```python
top_countries = df['job_country'].value_counts().head(6).index
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']
```

```python
job_salary_country = df.pivot_table(values = 'salary_year_avg', index= 'job_country', columns='job_title_short', aggfunc='median')
```

```python
job_salary_country = job_salary_country.loc[top_countries]
job_salary_country = job_salary_country[job_titles]
job_salary_country.plot(kind='bar')
plt.title('Median Salary for jobs per country ')
plt.ylabel('Median Salary')
plt.xlabel('')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```
